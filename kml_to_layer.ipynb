```python
def transform_single_value(value):
    try:
        return int(value)
    except (ValueError, TypeError):
        try:
            value = value.replace(',', '.')
            return float(value)
        except (ValueError, TypeError, AttributeError):
            try:
                return value
            except Exception as e:
                print(e)

def transform_all_values(values):
    """
    @transform all values from dataset
    @parameters:
        - values: dataset (list of lists) \n
    @return:
        - new_values: transformed dataset (list of lists)
    """
    #trata os dados de value
    new_values = []
    for list in values:
        newvalue = []
        for value in list:
            newvalue.append(transform_value(value))
        new_values.append(newvalue)
    new_values = transform_null_values(new_values)
    return new_values

def transform_null_values(values):
    """
    @converts string values '<Null>' to None Python object
    @parameters:
        - values: dataset (list of lists) \n
    @return:
        - new_list_values: transformed dataset (list of lists)
    """
    new_list_values = []
    for list_value in values:
        new_list = []
        for value in list_value:
            if value == '<Null>':
                value = None
            else:
                pass
            new_list.append(value)
        new_list_values.append(new_list)
    return new_list_values
```

```python
def get_infos_from_kml_popupinfo(path, fields, tag='td'):
    """
    @unpack 'PopupInfo' field collecting infos
    @parameters:
        - path: feature path
        - fields: feature's field List
        - tag: html tag used to unpack informations, 'td' as default \n
    @return:
        - tuple containing:
            - fields_list: feature's field list collected from PopupInfo field
            - values_list: values list of fields from fields_list
    """
    import re

    dict_src_names = {}

    i=0
    for name in fields:
        dict_src_names[name] = i
        i+=1

    cursor = arcpy.da.SearchCursor(path, fields)

    reg_str = "<" + tag + ">(.*?)</" + tag + ">"

    fields_list = []
    values_list = []
    for row in cursor:
        list_info = re.findall(reg_str, row[dict_src_names['PopupInfo']])[1:]
        print(list_info)
        field = []
        value = []
        field.append('Shape')
        value.append(row[dict_src_names['Shape']])
        for i, info in enumerate(list_info):
            if i % 2 == 0:
                field.append(info)
            else:
                value.append(info)
        fields_list.append(field)
        values_list.append(value)

    del cursor

    return (fields_list, values_list)
```
